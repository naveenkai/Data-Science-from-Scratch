## Types of Data

### 1. Numerical data - quantitative data (measured) - height, weight
### 2. Categorical data - categories - gender, nationality, color
### 3. Time series data - data points collected at regular intervals of time - stock prices, wether data, traffic data
### 4. Text data - unstructured data - words, sentences, emails, social media posts, reviews
### 5. Image data - photos, MRI scans, satellite images
### 6. Audio data - music, speech and sound effects

### 3. XML (eXtensible Markup Language)

```python
import xml.etree.ElementTree as ET
tree = ET.parse('BloodImage.xml')     ## parse - divide into grammatical parts
root = tree.getroot()

for child in root:
    print(child.tag, child.attrib)
```
